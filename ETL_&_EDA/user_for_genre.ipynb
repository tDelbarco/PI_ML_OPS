```python
import pandas as pd

# DataFrame original (reemplazar con tu DataFrame real)
data_original = {
    'genres': [None, None, ['Casual', 'Indie', 'Simulation', 'Strategy'], ['Casual', 'Indie', 'Strategy'], ['Indie', 'Racing', 'Simulation'], ['Casual', 'Indie'], None],
    'app_name': ['None', 'None', 'Colony On Mars', 'LOGistICAL: South Africa', 'Russian Roads', 'EXIT 2 - Directions', 'Maze Run VR'],
    'title': ['None', 'None', 'Colony On Mars', 'LOGistICAL: South Africa', 'Russian Roads', 'EXIT 2 - Directions', 'Maze Run VR'],
    'url': [None, None, 'http://store.steampowered.com/app/773640/Colony_On_Mars/', 'http://store.steampowered.com/app/733530/LOGistICAL_South_Africa/', 'http://store.steampowered.com/app/610660/Russian_Roads/', 'http://store.steampowered.com/app/658870/EXIT_2_Directions/', 'http://store.steampowered.com/app/681550/Maze_Run_VR/'],
    'reviews_url': [None, None, 'http://steamcommunity.com/app/773640/reviews/?browsefilter=mostrecent', 'http://steamcommunity.com/app/733530/reviews/?browsefilter=mostrecent', 'http://steamcommunity.com/app/610660/reviews/?browsefilter=mostrecent', 'http://steamcommunity.com/app/658870/reviews/?browsefilter=mostrecent', 'http://steamcommunity.com/app/681550/reviews/?browsefilter=mostrecent'],
    'price': [None, None, 1.99, 4.99, 1.99, 4.99, 4.99],
    'id': [None, None, 773640.0, 733530.0, 610660.0, 658870.0, 681550.0]
}

df_original = pd.DataFrame(data_original)

# Bucle para procesar la columna 'items'
for index, row in df_original.iterrows():
    items_df = pd.json_normalize(row['items'], sep='_') if 'items' in row and row['items'] else pd.DataFrame()
    
    # Merge con el DataFrame original
    merged_df = pd.merge(df_original, items_df, left_index=True, right_index=True, suffixes=('', '_items'), how='left')

    # Desanidar o split en la columna de géneros
    genres_df = pd.DataFrame(merged_df['genres_items'].to_list(), columns=['genre'])
    
    # Realizar el merge con el DataFrame original nuevamente
    final_df = pd.merge(merged_df, genres_df, left_index=True, right_index=True, suffixes=('', '_genres'), how='left')

    # Agrupar y sumar por año y género
    result_df = final_df.groupby(['year', 'genre'])['playtime_forever'].sum().reset_index()

    # Obtener el género con mayor horas jugadas por año
    max_playtime_genre = result_df.loc[result_df.groupby('year')['playtime_forever'].idxmax()]

    # Mostrar los resultados (puedes adaptar esto según tus necesidades)
    print(f"\nResultados para el índice {index}:\n{max_playtime_genre}")

    # Puedes almacenar o procesar más los resultados según tus necesidades
```
